```python
import math
import meep as mp
import numpy as np
from meep import mpb

num_bands = 30

k_points = [mp.Vector3(-0.5),          
            mp.Vector3(0),       
            mp.Vector3(0.5),  
           ]          

k_points = mp.interpolate(4, k_points)

dA = 0.8
dB = 0.2
epsilonA = 12
epsilonB = 3.7
geometry = []
def BlockAB(N, center):
        geometry = []
        for i in range(N):
            geometry = np.append(geometry, mp.Block(mp.Vector3(dA,mp.inf,mp.inf),
                         center=mp.Vector3(x=center-(dA+dB)*N/2 + (dA+dB)*i + dA/2),
                         material=mp.Medium(epsilon=epsilonA)))
            geometry = np.append(geometry, mp.Block(mp.Vector3(dB,mp.inf,mp.inf),
                         center=mp.Vector3(x=center-(dA+dB)*N/2 + (dA+dB)*i + dA + dB/2),
                         material=mp.Medium(epsilon=epsilonB)))
        geometry = geometry.tolist()
        return geometry
def BlockBA(N, center):
        geometry = []
        
        for i in range(N):
            geometry = np.append(geometry, mp.Block(mp.Vector3(dB,mp.inf,mp.inf),
                         center=mp.Vector3(x=center-(dA+dB)*N/2 + (dA+dB)*i + dB/2),
                         material=mp.Medium(epsilon=epsilonB)))
            geometry = np.append(geometry, mp.Block(mp.Vector3(dA,mp.inf,mp.inf),
                         center=mp.Vector3(x=center-(dA+dB)*N/2 + (dA+dB)*i + dB + dA/2),
                         material=mp.Medium(epsilon=epsilonA)))
        geometry = geometry.tolist()
        return geometry
Np = 2 ## number of repeating
geometry = sum([BlockAB(Np,Np*(dA+dB)/2),BlockBA(Np,Np*(dA+dB)/2)],[])
#for i in range (Np):
    #geometry.append(mp.Block(mp.Vector3(0.2),material= mp.Medium(epsilon=13)))

resolution = 32
geometry_lattice = mp.Lattice(size=mp.Vector3(1))
ms = mpb.ModeSolver(num_bands=num_bands,
                    k_points=k_points,
                    geometry=geometry,
                    geometry_lattice=geometry_lattice,
                    resolution=resolution)
ms.run_te()
```

```python
import matplotlib.pyplot as plt

freqs = ms.all_freqs*3*10**14
gaps = ms.gap_list

# DPI
dpi = 100  # 
fig, ax = plt.subplots(dpi=dpi)

x = range(len(freqs))
for xz, tz in zip(x, freqs):
    ax.scatter([xz] * len(tz), tz, color='red')
ax.plot(freqs, color='red')
ax.set_ylim([1.5*10**14, 2.4*10**14])
ax.set_xlim([x[0], x[-1]])
for gap in gaps:
    if gap[0] > 1:
        ax.fill_between(x, gap[1]*3*10**14, gap[2]*3*10**14, color='red', alpha=0.2)

points_in_between = (len(freqs) - 4) / 3
tick_locs = [i * points_in_between + i for i in range(4)]
transposed_array = np.transpose(ax)
ax.set_ylabel('Frequency', size=22)
ax.set_xlabel('k', size=22)
ax.grid(True)
plt.rcParams.update({'font.size': 20})
#plt.ylim(0.5,0.8)
# 

plt.savefig('bg_13_1_2')

plt.show()
```
