```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
#5AB 28
```

```python
def quadratic_function(t, a, b, tau1, tau2 , omega ,phi):
    return a * (1-np.exp((-t)/tau1)) + b * np.exp((-t)/tau2)*np.cos(omega*t+phi)
```
